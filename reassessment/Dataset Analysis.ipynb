```python
# Read the dataset from the txt file
mm_rainfall_data = pd.read_csv("./datasets/UK Met Office Public Data - mm Rainfall Series.txt", delim_whitespace=True, skiprows=5)
# Set the 'Year' column as the index
mm_rainfall_data.set_index('year', inplace=True)

mm_rainfall_data
```

```python
mm_rainfall_data.columns
```

```python
# Reset index to make 'Year' a column
mm_rainfall = mm_rainfall_data.reset_index()

# Convert the dataframe into long format
mm_rainfall_melted = pd.melt(mm_rainfall, id_vars='year', value_vars=['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec'])
# Create Date from year value and month 
mm_rainfall_melted['Date'] = pd.to_datetime(mm_rainfall_melted['year'].astype(str) + '-' + mm_rainfall_melted['variable'], format='%Y-%b')
mm_rainfall = mm_rainfall_melted.drop(columns=['year', 'variable'])
# Set the Date as index
mm_rainfall.set_index('Date',inplace=True)
# Rename the column to an unique name
mm_rainfall.columns = ['mm_rainfall']
# Sort DataFrame by Date
mm_rainfall = mm_rainfall.sort_index()
mm_rainfall
```
